```python
from web_email_scrapper import get_company_email_format
import re, time
import pandas as pd
```

```python
# Remove newline characters from each line
lines  = ["Google","Amazon","Microsoft","Facebook","Apple","Netflix","Adobe","Intel","Tesla","IBM"]

```

```python

# Start the timer
start_time = time.time()
all_companies = {}
cnt =0
for i in lines[:2]:
  print (cnt, ":", i)
  email_format, results = get_company_email_format(i)

  # Adding key as company name
  if results != "NA":
    company_name = i
    newdict = {company_name: results}
    # Print the result
    all_companies.update(newdict)
    cnt = cnt+1
    time.sleep(1)  # Sleep for 1 second to avoid rate limiting

# End the timer
end_time = time.time()
execution_time = end_time - start_time
print(f"Execution Time: {execution_time:.4f} seconds")
```

```python


# Patterns for email structure and accuracy
email_pattern = re.compile(r"(\[first_initial\]\[last\]|first \[1 letter\] \+ last|FLast|first\.last)")
accuracy_pattern = re.compile(r"(\d+(\.\d+)?%)")

# Dictionary to store results
company_results = {}

# Process each company
for company, records in all_companies.items():
    extracted_details = []
    for record in records:
        email_structure = email_pattern.search(record['snippet'])
        accuracy = accuracy_pattern.search(record['snippet'])
        extracted_details.append({
            'link': record['link'],
            'email_structure': email_structure.group(0) if email_structure else 'N/A',
            'accuracy': accuracy.group(0) if accuracy else 'N/A'
        })
    company_results[company] = extracted_details

# Output the dictionary
print(company_results)
```

```python
# Flatten the dictionary into a list of records
flattened_data = []
for company, records in company_results.items():
    for record in records:
        flattened_data.append({
            'Company': company,
            'Accuracy': record['accuracy'],
            'Email Structure': record['email_structure'],
            'Link': record['link']
        })

# Convert the list into a DataFrame
df = pd.DataFrame(flattened_data)

# Print the DataFrame
print(df.head(50))
```

```python
import requests
from lxml.html import fromstring
from itertools import cycle
import traceback

# Function to fetch proxies from the website
def to_get_proxies():
    # Website to get free proxies
    url = 'https://free-proxy-list.net/'

    response = requests.get(url)

    parser = fromstring(response.text)
    # Using a set to avoid duplicate IP entries
    proxies = set()

    # Parsing the proxy list (get the first 10 entries)
    for i in parser.xpath('//tbody/tr')[:10]:
        # To check if the proxy supports HTTPS
        if i.xpath('.//td[7][contains(text(),"yes")]'):
            # Grab the proxy IP and corresponding PORT
            proxy = ":".join([i.xpath('.//td[1]/text()')[0], i.xpath('.//td[2]/text()')[0]])
            proxies.add(proxy)

    return proxies

# Fetch the proxies
proxies = to_get_proxies()

# Print the proxies
print("Proxies fetched:", proxies)


# Rotate through the list of proxies
proxyPool = cycle(proxies)

# URL to scrape (you can insert the desired URL here)
url = 'https://httpbin.org/ip'  # Example URL for testing, replace with your target URL

# Iterate and make requests with rotated proxies
for i in range(1, 11):
    # Get a proxy from the pool
    proxy = next(proxyPool)
    print(f"Request #{i} using proxy: {proxy}")

    try:
        # Send GET request with the proxy
        response = requests.get(url, proxies={"http": proxy, "https": proxy}, timeout=5)
        print("Response status code:", response.status_code)
        print("Response content:", response.json())

    except requests.exceptions.RequestException as e:
        # Handle specific exceptions
        print(f"Skipping proxy {proxy}. Connection error: {e}")
        # Optionally, you can log the exception details or retry with another proxy
        traceback.print_exc()
```
