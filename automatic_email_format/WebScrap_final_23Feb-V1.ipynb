```python
import pandas as pd
import difflib
import re

def extract_domain(email_pattern):
    """Extract domain from the email pattern."""
    match = re.search(r'@([\w.-]+)', str(email_pattern))  # Convert to string (handle NaN)
    return match.group(1) if match else None

def similarity_score(company, domain):
    """Compare similarity between company name and domain."""
    company_clean = re.sub(r'[^a-zA-Z0-9]', '', str(company)).lower()
    domain_clean = re.sub(r'\..*$', '', str(domain)).lower()  # Remove TLD (.com, .edu)

    score = difflib.SequenceMatcher(None, company_clean, domain_clean).ratio()

    if score > 0.8:
        return "High"
    elif score > 0.4:
        return "Medium"
    else:
        return "Low"

def get_best_email(company, email_list):
    """Select the best email and its similarity score (High, Medium, Low)."""
    best_score = None
    best_email = None
    best_flag = None

    for email in email_list:
        domain = extract_domain(email)
        score = similarity_score(company, domain)
        
        if best_score is None or \
           (score == "High" and best_score != "High") or \
           (score == "Medium" and best_score not in ["High", "Medium"]) or \
           (score == "Low" and best_score == "Low"):
            best_score = score
            best_email = email
            best_flag = score  # Store the flag

    return best_email, best_flag

# Example dataframe
data = {
    'company': [
        'minnesota exteriors', 'minnesota exteriors', 'mit department of biological engineering',
        'lincoln foodservice', 'lincoln foodservice', 'usa bluebook', 'usa bluebook',
        'usa bluebook', 'connectivity standards alliance', 'connectivity standards alliance', 
        'south florida surgical specialists, llc'
    ],
    'email_list': [
        ['jane.doe@minnesotaexteriors.com'],
        ['jsmith@northernexteriorsmn.com'],
        ['jdoe@mit.edu'],
        ['jdoe@lincolnfp.com'],
        ['jane.doe@lincolnindustries.com'],
        ['jdoe@usabluebook.com'],
        ['John.Smith@usabluebook.com', 'JSmith@usabluebook.com'],
        ['jdoe@thebluebook.com'],
        ['jsmith@csa-iot.org'],
        ['jsmith@csa-iot.org'],
        ['jsmith@miamivipsurgery.com']
    ]
}

df = pd.DataFrame(data)
print (df)
# Apply function to get best email and similarity score
df[['best_email', 'match_flag']] = df.apply(lambda row: pd.Series(get_best_email(row['company'], row['email_list'])), axis=1)

print(df[['company', 'best_email', 'match_flag']])
```

```python
import os
import time
import pandas as pd

output_path = r'email_creator_app/files/missing_emails/EW-ND-10Mar-_fo_Missing_file2025-03-10 14-33-02_missing.csv'
checkpoint_path = 'progress_checkpoint.txt'  # File to track last completed batch
output_path = output_path[:5]
# Read the CSV file
missing_data = pd.read_csv(output_path)

queries = [str(i) for i in missing_data['Company']]
print(f"Total queries: {len(queries)}")

# Define batch size
BATCH_SIZE = 2590

# Get total number of batches
num_batches = len(queries) // BATCH_SIZE
remainder = len(queries) % BATCH_SIZE

# Determine where to start
last_completed_batch = -1  # Default to start from the beginning
if os.path.exists(checkpoint_path):
    with open(checkpoint_path, 'r') as f:
        last_completed_batch = int(f.read().strip())

print(f"Resuming from batch: {last_completed_batch + 1}")

# Iterate over batches
for batch_idx in range(last_completed_batch + 1, num_batches + (1 if remainder > 0 else 0)):
    start_index = batch_idx * BATCH_SIZE
    end_index = min(start_index + BATCH_SIZE, len(queries))
    batch = queries[start_index:end_index]

    print(f"Processing Batch {start_index}_{end_index}")
    namefile = f"Batch {start_index}_{end_index}__SCRAPPER_"

    try:
        df_raw = scrapper_manager(batch, namefile)
        df_raw.to_csv(f"FinalDataBatch{namefile}9madddrch.csv")
        print(f"Batch {start_index}_{end_index} saved. Shape: {df_raw.shape}")

        # Update checkpoint
        with open(checkpoint_path, 'w') as f:
            f.write(str(batch_idx))

    except Exception as e:
        print(f"Error in batch {batch_idx}: {e}")
        break  # Stop the loop so the issue can be fixed

    time.sleep(30)  # Pause between batches
```
