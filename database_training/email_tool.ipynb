```python
import pandas as pd

# Sample data for DataFrame A (company name and email format)
data_a = {
    'CompanyName': ['10th special forces group', '10th sfg', '10th sfga'],
    'EmailFormat': ['{LastName}{FirstName}1@soc.mil', '{LastName}{FirstName}1@soc.mil', '{LastName}{FirstName}1@soc.mil']
}

# Sample data for DataFrame B (first name, last name, company name)
data_b = {
    'FirstName': ['John', 'Jane'],
    'LastName': ['Doe', 'Smith'],
    'CompanyName': ['10th sfg', '10th special forces group']
}

# Convert dictionaries to dataframes
df_a = pd.DataFrame(data_a)
df_b = pd.DataFrame(data_b)

# Function to generate email from first name, last name, and email format
def generate_email(first_name, last_name, email_format):
    # Format the email based on the provided format
    return email_format.format(FirstName=first_name, LastName=last_name)

# Merge DataFrame B with DataFrame A on 'CompanyName' to get the corresponding email format
df_b = df_b.merge(df_a, on='CompanyName', how='left')
print(df_b.head())
# Generate the email column in DataFrame B using the email format from DataFrame A
df_b['Email'] = df_b.apply(lambda row: generate_email(row['FirstName'], row['LastName'], row['EmailFormat']), axis=1)

# Output the result
print(df_b.head())
```

```python
import pandas as pd
import pickle
file_path = r'/Users/sumanverma/Documents/Work/Email_tool/company_database/output/Colorado Springs, CO Seggrigation __2024-12-29_output.csv'
# Step 1: Read the CSV file
csv_file = file_path# Replace with your actual CSV file path
data = pd.read_csv(csv_file)

# Step 2: Create the dictionary
email_dict = {}
for _, row in data.iterrows():
    company = row['company'] # Remove extra spaces if any
    email_format = row['email pattern'] # Remove extra spaces if any
    if company not in email_dict:
        email_dict[company] = set()
    email_dict[company].add(email_format)

# Convert sets back to lists for easier serialization
email_dict = {key: list(value) for key, value in email_dict.items()}

# Step 3: Save the dictionary as a pickle file
pickle_file = r"/Users/sumanverma/Documents/Work/Email_tool/email_patterns.pkl"
with open(pickle_file, "wb") as file:
    pickle.dump(email_dict, file)

print(f"Pickle file created: {pickle_file}")
```
